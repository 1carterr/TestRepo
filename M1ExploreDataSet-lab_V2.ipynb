## Find out the number of rows and columns


Start by exploring the numbers of rows and columns of data in the dataset.


Print the number of rows in the dataset.


```python
## Write your code here
import pandas as pd

# Load the dataset directly from the URL
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(file_path)

# Display top 5 rows
print("Top 5 rows of the dataset:")
print(df.head())

# Display top 5 columns (column names)
print("\nTop 5 columns of the dataset:")
print(df.columns[:5])

# Check for duplicate rows
duplicate_rows = df.duplicated()
print(f"\nNumber of duplicate rows: {duplicate_rows.sum()}")

# Display first 5 duplicate rows, if any
if duplicate_rows.sum() > 0:
    print("\nFirst 5 duplicate rows:")
    print(df[duplicate_rows].head())

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values per column:")
print(missing_values)
```
