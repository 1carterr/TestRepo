### Section 1: Handling Duplicates
##### Task 1: Identify and remove duplicate rows.


```python
## Write your code here
import pandas as pd

# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Task 1: Identify duplicate rows
duplicate_rows = df.duplicated()

# Print the number of duplicate rows
print(f"Number of duplicate rows: {duplicate_rows.sum()}")

# Display the first few duplicate rows
print("\nFirst few duplicate rows:")
print(df[duplicate_rows].head())

# Remove duplicate rows
df_cleaned = df.drop_duplicates()

# Verify by checking the number of duplicate rows after removal
duplicate_rows_after_removal = df_cleaned.duplicated().sum()

# Print the number of duplicate rows after removal
print(f"\nNumber of duplicate rows after removal: {duplicate_rows_after_removal}")
```

##### Task 5: Normalize ConvertedCompYearly using Min-Max Scaling.
Min-Max Scaling brings all values in a column to a 0-1 range, making it useful for comparing data across different scales. Here, you will apply Min-Max normalization to the ConvertedCompYearly column, creating a new column ConvertedCompYearly_MinMax with normalized values.


```python
## Write your code here
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Task 5: Normalize ConvertedCompYearly using Min-Max Scaling
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Ensure that the ConvertedCompYearly column is numeric (in case of non-numeric data)
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Apply Min-Max scaling to the 'ConvertedCompYearly' column
df['ConvertedCompYearly_MinMax'] = scaler.fit_transform(df[['ConvertedCompYearly']])

# Display the first few rows to check the new column
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax']].head())
```
