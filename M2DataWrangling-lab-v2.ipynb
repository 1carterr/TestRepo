<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
import pandas as pd

# Define the URL of the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_url)

# Task 3.2: Standardize entries in 'Country' and 'EdLevel' columns

# Standardizing the 'Country' column
country_mapping = {
    'USA': 'United States',
    'US': 'United States',
    'United States of America': 'United States',
    'UK': 'United Kingdom',
    'England': 'United Kingdom',
    'Canada': 'Canada',
    'India': 'India'
    # Add more mappings if needed
}

# Map the values in the 'Country' column to standardized format
df['Country'] = df['Country'].map(country_mapping).fillna(df['Country'])

# Standardizing the 'EdLevel' column
edlevel_mapping = {
    "Master's": "Master's",
    "MSc": "Master's",
    "MS": "Master's",
    "Bachelors": "Bachelor's",
    "High School": "High School",
    "PhD": "PhD"
    # Add more mappings if needed
}

# Map the values in the 'EdLevel' column to standardized format
df['EdLevel'] = df['EdLevel'].map(edlevel_mapping).fillna(df['EdLevel'])

# Display the unique values of 'Country' and 'EdLevel' after standardization
print("\nUnique countries after standardization:")
print(df['Country'].unique())

print("\nUnique education levels after standardization:")
print(df['EdLevel'].unique())
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
import pandas as pd

# Define the URL of the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_url)

# Task 5.2: Impute missing values in 'ConvertedCompYearly' with the mean or median

# Impute with the mean of the column
mean_value = df['ConvertedCompYearly'].mean()

# Alternatively, you can use median_value if preferred:
# median_value = df['ConvertedCompYearly'].median()

# Impute missing values with the mean
df['ConvertedCompYearly'].fillna(mean_value, inplace=True)

# Verify the imputation by checking for missing values in the column
missing_after_imputation = df['ConvertedCompYearly'].isnull().sum()

print(f"Missing values in 'ConvertedCompYearly' after imputation: {missing_after_imputation}")
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
import pandas as pd

# Define the URL of the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_url)

# Task 5.3: Impute missing values in 'RemoteWork' with the most frequent value

# Find the most frequent value (mode) in the 'RemoteWork' column
most_frequent_remote_work = df['RemoteWork'].mode()[0]

# Impute missing values with the most frequent value
df['RemoteWork'].fillna(most_frequent_remote_work, inplace=True)

# Verify the imputation by checking for missing values in the column
missing_after_imputation = df['RemoteWork'].isnull().sum()

print(f"Missing values in 'RemoteWork' after imputation: {missing_after_imputation}")
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
import pandas as pd

# Define the URL of the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_url)

# Task 6.1: Apply Min-Max Scaling to normalize 'ConvertedCompYearly' manually using pandas

# Min-Max Scaling formula
min_value = df['ConvertedCompYearly'].min()
max_value = df['ConvertedCompYearly'].max()

# Normalize the 'ConvertedCompYearly' column
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - min_value) / (max_value - min_value)

# Display the first few rows to verify the scaling
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import pandas as pd
import numpy as np

# Define the URL of the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_url)

# Task 6.2: Log-transform 'ConvertedCompYearly' column to reduce skewness

# Apply log transformation (add 1 to avoid log(0))
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)

# Display the first few rows to verify the transformation
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
import pandas as pd

# Define the URL of the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(dataset_url)

# Task 7.1: Create 'ExperienceLevel' column based on 'YearsCodePro'

# Define experience level categories based on 'YearsCodePro'
bins = [0, 2, 5, 10, float('inf')]  # Bin edges
labels = ['Entry-level', 'Mid-level', 'Senior-level', 'Expert-level']  # Corresponding labels

# Create the 'ExperienceLevel' column using pd.cut()
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels, right=False)

# Display the first few rows to verify the new column
print(df[['YearsCodePro', 'ExperienceLevel']].head())
```
