## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
# For sending HTTP requests
import requests

# For parsing HTML content
from bs4 import BeautifulSoup

# Optional: For storing and working with tabular data
import pandas as pd
```

Download the webpage at the url


```python
#your code goes here
import requests

# URL of the webpage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

# Send HTTP request
response = requests.get(url)

# Save the content to a local HTML file
with open("Programming_Languages.html", "w", encoding="utf-8") as file:
    file.write(response.text)

print("Webpage downloaded successfully.")
```

Create a soup object


```python
#your code goes here
from bs4 import BeautifulSoup

# Open the downloaded HTML file
with open("Programming_Languages.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, "html.parser")

print("Soup object created successfully.")
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
# Find the table
table = soup.find("table")

# Create empty lists to store the data
languages = []
salaries = []

# Loop through table rows (skip the header row)
for row in table.find_all("tr")[1:]:
    cols = row.find_all("td")
    
    if len(cols) >= 2:
        language = cols[0].get_text(strip=True)
        salary = cols[1].get_text(strip=True)
        
        languages.append(language)
        salaries.append(salary)

# Display the scraped data
for lang, sal in zip(languages, salaries):
    print(f"Language: {lang}, Average Annual Salary: {sal}")
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
import pandas as pd

# Create a DataFrame from the scraped lists
df = pd.DataFrame({
    "Programming Language": languages,
    "Average Annual Salary": salaries
})

# Save to CSV file
df.to_csv("popular-languages.csv", index=False)

print("Data saved successfully to popular-languages.csv")
```
